```python
import pandas as pd




```

```python
df = pd.read_csv("tickets.csv")
df.head()
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z ]', '', text)
    text = text.strip()
    return text

df['clean_message'] = df['Ticket Description'].apply(clean_text)

```

```python
def classify_issue(text):
    if 'login' in text or 'password' in text or 'account' in text:
        return 'LOGIN'
    elif 'payment' in text or 'card' in text or 'transaction' in text:
        return 'PAYMENT'
    elif 'delivery' in text or 'delay' in text or 'shipment' in text:
        return 'DELIVERY'
    elif 'refund' in text or 'return' in text:
        return 'REFUND'
    elif 'bug' in text or 'error' in text or 'failed' in text or 'not working' in text:
        return 'BUG'
    else:
        return 'GENERAL'

df['issue_type'] = df['clean_message'].apply(classify_issue)
```

```python
df[['Ticket Description', 'issue_type']].head(10)
```

```python
def assign_priority(text):
    if 'urgent' in text or 'not working' in text or 'immediately' in text:
        return 'P0'   # Critical
    elif 'failed' in text or 'error' in text or 'unable' in text:
        return 'P1'   # High
    elif 'delay' in text or 'late' in text:
        return 'P2'   # Medium
    else:
        return 'P3'   # Low

df['priority'] = df['clean_message'].apply(assign_priority)
```

```python
df[['Ticket Description', 'priority']].head()
```

```python
def sla_hours(priority):
    if priority == 'P0':
        return 2      # Critical
    elif priority == 'P1':
        return 6      # High
    elif priority == 'P2':
        return 24     # Medium
    else:
        return 48     # Low

df['sla_hours'] = df['priority'].apply(sla_hours)
```

```python
df[['priority', 'sla_hours']].head(10)
```

```python
df.to_csv("final_ticket_report.csv", index=False)
print("Final report generated successfully!")
```
